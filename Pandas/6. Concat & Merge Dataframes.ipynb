```python
temp_df = pd.DataFrame({
    "city" :['New York','Chicago','Orlando','Baltimore'],
    "temperature" : [21,14,32,35]
})
temp_df
```

```python
humid_df = humidity_df = pd.DataFrame({
    "city" :['Chicago','Orlando','San Fransisco'],
    "humidity" : [65,68,70]
})
humid_df
```

```python
df2 = pd.merge(temp_df,humid_df) # inner join or intersection : only common elements are taken
df2
```

```python
df3 = pd.merge(temp_df,humid_df,how='outer') # union
df3
```

```python
df4 = pd.merge(temp_df,humid_df,how='inner') # inner is the default
df4
```

```python
df5 = pd.merge(temp_df,humid_df,how="left") # take intersection and elementsthat are in left dataframe
df5
```

```python
df6 = pd.merge(temp_df,humid_df,how="right") # right join
df6
```

```python
df3 = pd.merge(temp_df,humid_df,how='outer',indicator=True) # to see from which dataframe Data is coming
df3
```

```python
df1 = pd.DataFrame({
    "city" :['Chicago','Orlando','New york','Baltimore'],
    "humidity" : [65,68,70,75],
    "temperature" : [21,14,32,38]
})
df1
```
