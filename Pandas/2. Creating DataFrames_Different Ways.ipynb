```python
import pandas as pd
```

### Different ways of creating dataframes
- 1) Using CSV
- 2) Using Excel
- 3) From Python dictionary
- 4) From list of tuples
- 5) From list of dictionaries

```python
# Using CSV
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/3_different_ways_of_creating_dataframe/weather_data.csv")
df
```

```python
# Using excel file
df = pd.read_excel("weather_data.xlsx")
df
```

```python
#  3. Creating dataframe from python dictionary
weather_data = {
    'day' : ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'temperature':[32,35,28,24,32,31],
    'windspeed':[6,7,2,7,4,2],
    'event':['Rain','Sunny','Snow','Snow','Rain','Sunny']
}

df = pd.DataFrame(weather_data)
df
```

```python
# 4. Using list of tuples

weather_data = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]

# we need to mention the column names using tuples list
df = pd.DataFrame(weather_data,columns=['day','temperature','windspeed','event'])
df
```

```python
# 5. Creating dataframe from list of dictionaries

weather_data = [
    {'day':'1/1/2017','temperature':32,'windspeed':6,'event':'Rain'},
    {'day':'1/2/2017','temperature':35,'windspeed':8,'event':'Snow'}
]

df= pd.DataFrame(weather_data)
df
```
