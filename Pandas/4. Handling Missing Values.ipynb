```python
# convert day column to date column

df = pd.read_csv(file_path, parse_dates=['day'])
df.head()
```

```python
type(df.day[0])
```

```python
# make day column as index of data frame
df.set_index('day',inplace=True)
df
```

# fillna()

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed' : 0,
    'event' : 'no event'
})
new_df
```

```python
new_df = df.fillna(method="ffill") # forward fill : carry forward previous days value
new_df
```

```python
new_df = df.fillna(method="bfill") # forward fill : replace with next days value
new_df
```

```python
new_df = df.fillna(method="ffill",limit=1) # copy only once
new_df
```

# interploate

```python
new_df = df.interpolate() # default is linear interpolation
new_df
```

```python
new_df = df.interpolate(method='quadratic')
new_df
```

```python
new_df = df.interpolate(method='time') # calculates time difference between the dates and replace the NAN accordingly
new_df
```

# dropna()

```python
new_df = df.dropna()
new_df
```

```python
# drop if a row has all missing values
new_df = df.dropna(how='all') # drop rows where every value is NaN
new_df
```

```python
# if atleast one not null value present in a row don't drop it
new_df = df.dropna(thresh=1)
new_df
```

# replace()

```python
df = pd.DataFrame(
{
    'day' : ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017','1/6/2017'],
    'temperature' : [32,-99999,28,-99999,32,31,34],
    'windspeed' : [6,7,-99999,7,-88888,2,5],
    'event' : ['Rain','Sunny','Snow',0,'Rain','Sunny',0]
})
df
```

```python
new_df = df.replace(-99999,np.NaN) # replace -99999 with NaN
new_df
```

```python
new_df = df.replace([-99999,-88888],np.NaN) # replace -99999,-88888 with NaN
new_df
```

```python
# replace values based on column name
new_df = df.replace(
{
    'temperature':-99999,
    'windspeed' : -88888,
    'event' : 0
}, np.NaN)

new_df
```

```python
df = pd.DataFrame(
{
    'day' : ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017','1/6/2017'],
    'temperature' : [32,-99999,28,-99999,32,31,34],
    'windspeed' : [6,7,-99999,7,-88888,2,5],
    'event' : ['Rain','Sunny','Snow','No event','Rain','Sunny','No event']
})
df
```

```python
new_df = df.replace({
    -99999 : np.NaN,
    -88888 :np.NaN,
    'No event' : 'Sunny'
})

new_df
```

```python
df = pd.DataFrame(
{
    'day' : ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017','1/6/2017'],
    'temperature' : ['32F',-99999,28,-99999,'32C',31,34],
    'windspeed' : ['6mph',7,-99999,'7mph',-88888,2,5],
    'event' : ['Rain','Sunny','Snow','No event','Rain','Sunny','No event']
})
df
```

```python
new_df = df.replace('[A-Za-z]','',regex=True) # replace anythong bewtwwen A_Za-z with blank
new_df
```

```python
new_df = df.replace({
                     'temperature':'[A-Za-z]',
                     'windspeed':'[A-Za-z]'
                    },'', regex=True)


new_df
```
