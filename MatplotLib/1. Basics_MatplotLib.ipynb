```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
# a basic 2D plot
plt.plot(x,y)
```

```python
# Adding lables and title
plt.plot(x,y,color='green',linewidth=5,linestyle='dotted')
plt.xlabel('Day')
plt.ylabel('Temparature')
plt.title('Weather')
```

```python
# specifying the line color and style
plt.plot(x,y,'g--')
```

```python
plt.plot(x,y,'b-.')
```

```python
plt.plot(x,y,'--r+')
```

```python
plt.plot(x,y,'yD') # yellow color Diamond marker
```

```python
plt.plot(x,y,color='yellow',marker='D',linestyle='') # yellow color Diamond marker
```

```python
# control marker size
plt.plot(x,y,color='red',marker='+',linestyle='dashed',markersize=20)
```

```python
# controlling line trasperency using alpha, alpha takes values between 0 to 1

plt.plot(x,y,color='purple',alpha=0.5)
```

```python
plt.plot(x,y,color='purple')
```

# axis lables, legend and grid

```python
days = [1,2,3,4,5,6,7]
max_temp = [50, 51, 52, 48, 47, 49, 46]
min_temp = [43,42,40,44,33,35,37]
avg_temp = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("weather")

plt.plot(days,max_temp,label="Max")
plt.plot(days,min_temp,label="Min")
plt.plot(days,avg_temp,label="Avg")

plt.legend(loc="upper right")
```

```python
plt.xlabel('Days',fontsize=15)
plt.ylabel('Temperature',fontsize=15)
plt.title("weather",fontsize=25)

plt.plot(days,max_temp,label="Max")
plt.plot(days,min_temp,label="Min")
plt.plot(days,avg_temp,label="Avg")

plt.legend(loc = "best",shadow = True, fontsize="large")
```

```python
plt.xlabel('Days',fontsize=15)
plt.ylabel('Temperature',fontsize=15)
plt.title("weather",fontsize=25)

plt.plot(days,max_temp,label="Max")
plt.plot(days,min_temp,label="Min")
plt.plot(days,avg_temp,label="Avg")

plt.legend(loc = "best",shadow = True, fontsize="medium")

plt.grid()
```
