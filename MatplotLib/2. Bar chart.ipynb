```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
company = ['Google','Amazon','Microsoft','FaceBook']
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
plt.xlabel('company')
plt.ylabel('Revenue (bln)')
plt.title(" USA Tech Stocks",fontsize=15)
plt.bar(company,revenue,label='Revenue')
plt.legend()
```

```python
# plot profit on same chart : side by side bar chart (to compare revenue and profit)
plt.xlabel('company')
plt.ylabel('Revenue (bln)')
plt.title(" USA Tech Stocks",fontsize=15)
plt.bar(company,revenue,label='Revenue')
plt.bar(company, profit, label='Profit')
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company) # takes to arguments positions and labels
#0 will be gOOGLE, 1 will be Amazon etc
```

```python
plt.xticks(xpos,company)

plt.bar(xpos,revenue)
```

```python
plt.xlabel('company')
plt.ylabel('Revenue (bln)')
plt.title(" USA Tech Stocks",fontsize=15)

plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,label='Revenue',width=0.4)
plt.bar(xpos+0.2, profit, label='Profit',width=0.4)
plt.legend()
```

```python
plt.ylabel('company')
plt.xlabel('Revenue (bln)')
plt.title(" USA Tech Stocks",fontsize=15)

plt.yticks(xpos,company)
plt.barh(xpos-0.2,revenue,label='Revenue')
plt.barh(xpos+0.2, profit, label='Profit')
plt.legend()
```
