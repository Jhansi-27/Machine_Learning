```python
import matplotlib.pyplot as plt
```

```python
exp_vals = [750, 250, 100, 50, 120]
exp_labels = ['Home Rent', "Food", "Internet", "Car", "Water/Electricity"]
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1.5)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1.75,autopct="%0.2f%%", shadow=True)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1.95,autopct="%0.2f%%", explode=[0,0.2,0.2,0,0])
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1.95,autopct="%0.2f%%", explode=[0,0.2,0.2,0,0],startangle=180)
plt.show()
```
