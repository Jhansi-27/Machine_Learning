```python
#Histograms are barcharts where y-axis contains frequency.
```

```python
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113, 85, 90, 150, 149, 89, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar) # by default 10 bins are created
```

```python
plt.hist(blood_sugar,bins=3)
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.95) # relative width of a bar compared to a bin size
```

```python
# we want to bin poatients into 3 categories
# 80-100 : Normal
# 100-125 : pre-diabetic
# 125-150 : diabetic

# we can do this by specifying the values in bin

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95, color='g')
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95, color='g',histtype='step')
```

```python
blood_sugar_men= [113, 85, 90, 150, 149, 89, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89,120,133, 84, 69, 89, 79,120,100,95]


plt.xlabel('sugar_range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis of men and women')

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],
         rwidth=0.95,color=['g','orange'],
         label=['men','women'])

plt.legend()
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('sugar_range',fontsize=15)
plt.ylabel('Total no of patients',fontsize=15)
plt.title('Blood sugar analysis',fontsize=20)

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],
         rwidth=0.95,color=['g','orange'],
         label=['men','women'],
         orientation='horizontal')

plt.legend(shadow=True,fontsize='large')
```
